### Clean Data

### Price Index

```python
# Load data
price_index = pd.read_csv('data/raw/prices/price-index/18100205.csv')

# Filter to CMA rows
price_index = price_index[~price_index['GEO'].isin(['Canada', 'Atlantic Region', 'Newfoundland and Labrador',
                                                    'Prince Edward Island', 'Nova Scotia', 'New Brunswick',
                                                    'Quebec', 'Ontario', 'Prairie Region', 'Manitoba', 'Saskatchewan',
                                                    'Alberta', 'British Columbia'])]

# Filter to housing price rows
price_index = price_index[price_index['New housing price indexes'] == 'House only']

# Extract year and month from date
price_index['year'] = price_index['REF_DATE'].str[:4].astype(int)
price_index['month'] = price_index['REF_DATE'].str[-2:].map(MONTH_DICT)
price_index['REF_DATE'] = pd.to_datetime(price_index['REF_DATE'])
price_index['series_position'] = price_index.groupby('GEO')['REF_DATE'].rank(method='dense').astype(int)

# Set CMA ID to be equal to Saint John CMA ID for combined Saint John, Fredericton, and Moncton data
price_index.loc[price_index['DGUID'].isna(), 'DGUID'] = '2011S0503310'

# Replace CMA ID for St. John's
price_index.loc[price_index['DGUID'] == '2011S05031', 'DGUID'] = '2011S0503001'

# Remove rows with missing data
price_index = price_index[price_index['VALUE'].isna() == False]

# Rename columns
price_index = price_index.rename({'REF_DATE': 'ref_date', 'GEO': 'cma', 'DGUID': 'cma_id', 'VALUE': 'price_index'}, axis=1)

# Select columns of interest
price_index = price_index[['ref_date', 'series_position', 'month', 'year', 'cma', 'cma_id', 'price_index']]
```

### Population

```python
# Load data
population = pd.read_csv('data/raw/demographics/population/17100135.csv')

# Filter to CMAs of interest
population = population[population['DGUID'].isin(CMA_IDS)]

# Filter to categories of interest
population = population[population['Sex'] == 'Both sexes']
population = population[population['Age group'] == 'All ages']

# Create summed row for Saint John, Fredericton, and Moncton CMAs
nb_cmas = ['2016S0504320', '2016S0503305', '2016S0503310']
nb_population = population[population['DGUID'].isin(nb_cmas)]
summed_nb = nb_population.groupby('REF_DATE')['VALUE'].sum().reset_index()

# Append summed Saint John, Fredericton, and Moncton data and drop original rows
nb_rows = pd.DataFrame({
    'GEO': 'Saint John, Fredericton, and Moncton (CMA), New Brunswick',
    'DGUID': '2016S0503310',
    'REF_DATE': summed_nb['REF_DATE'],
    'VALUE': summed_nb['VALUE']
})
population = population[~population['DGUID'].isin(nb_cmas)]
population = pd.concat([population, nb_rows], ignore_index=True).sort_values(by='REF_DATE')

# Add column for month
population['month'] = 'july'

# Rename columns
population = population.rename({'REF_DATE': 'year', 'DGUID': 'cma_id', 'VALUE': 'population'}, axis=1)

# Filter to columns of interest
population = population[['year', 'month', 'cma_id', 'population']]
```

```python
# Load data
construction_investment = pd.read_csv('data/raw/supply/construction-investment/34100286.csv')

# Filter to CMAs of interest
construction_investment = construction_investment[construction_investment['DGUID'].isin(CMA_IDS)]

# Filter to categories of interest
structures = ['Total residential', 'Single', 'Row', 'Apartment']
work = ['New construction', 'Conversions', 'Conversions total', 'Renovations', 'Renovations total', 'Types of work, total']
construction_investment = construction_investment[construction_investment['Type of structure'].isin(structures)]
construction_investment = construction_investment[construction_investment['Type of work'].isin(work)]
construction_investment = construction_investment[construction_investment['Investment Value'] == 'Unadjusted - constant']

# Create new column for category
construction_investment['category'] = construction_investment['Type of work'] + ' ' + construction_investment['Type of structure']
construction_investment['category'] = construction_investment['category'].str.lower().str.replace(', total', '').str.replace(' ', '_')

# Pivot data from long to wide
index = ['REF_DATE', 'GEO', 'DGUID']
categories = construction_investment['category'].unique()
construction_investment = construction_investment.pivot(index=index, columns='category', values='VALUE')
construction_investment = construction_investment.reset_index()

# Remove empty columns
for col in categories:
    if sum(construction_investment[col].isna() == False) == 0 or sum(construction_investment[col]) == 0:
        construction_investment = construction_investment.drop(columns=col)

# Create summed row for Saint John, Fredericton, and Moncton CMAs
new_categories = ['new_construction_apartment', 'new_construction_row', 'new_construction_single',
                  'types_of_work_apartment', 'types_of_work_row', 'types_of_work_single', 'types_of_work_total_residential']
nb_cmas = ['2016S0504320', '2016S0503305', '2016S0503310']
nb_construction = construction_investment[construction_investment['DGUID'].isin(nb_cmas)]
summed_nb = nb_construction.groupby('REF_DATE')[new_categories].sum().reset_index()

# Append summed Saint John, Fredericton, and Moncton data and drop original rows
nb_rows = pd.DataFrame({
    'GEO': 'Saint John, Fredericton, and Moncton (CMA), New Brunswick',
    'DGUID': '2016S0503310',
    'REF_DATE': summed_nb['REF_DATE'],
    'new_construction_apartment': summed_nb['new_construction_apartment'],
    'new_construction_row': summed_nb['new_construction_row'],
    'new_construction_single': summed_nb['new_construction_single'],
    'types_of_work_apartment': summed_nb['types_of_work_apartment'],
    'types_of_work_row': summed_nb['types_of_work_row'],
    'types_of_work_single': summed_nb['types_of_work_single'],
    'types_of_work_total_residential': summed_nb['types_of_work_total_residential']
})
construction_investment = construction_investment[~construction_investment['DGUID'].isin(nb_cmas)]
construction_investment = pd.concat([construction_investment, nb_rows], ignore_index=True)

# Extract year and month from date
construction_investment['year'] = construction_investment['REF_DATE'].str[:4].astype(int)
construction_investment['month'] = construction_investment['REF_DATE'].str[-2:].map(MONTH_DICT)

# Rename columns
construction_investment = construction_investment.rename({'DGUID': 'cma_id'}, axis=1)

# Filter to columns of interest
construction_investment = construction_investment.drop(columns=['REF_DATE', 'GEO'])
```

### Combine Data

```python
# Define function to linearly interpolate data forcing intercept through first available data
def linearly_interpolate_before_start(data, var):
    
    # Identify minimum series position for which population data is available
    first_non_na_index = data[var].first_valid_index()
    first_non_na_series_position = data.loc[first_non_na_index, 'series_position']
    first_non_na_var = data.loc[first_non_na_index, var]
    
    # Drop rows with missing population values
    model_data = data.dropna(subset=[var])

    # Select first ten data points
    model_data = model_data[:10]
    
    # Extract features and target
    X = model_data['series_position'].values.reshape(-1, 1)
    y = model_data[var].values.reshape(-1, 1)
    
    # Fit linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Calculate the slope of the regression line
    slope = model.coef_[0][0]
    
    # Calculate the intercept to force the regression line through the desired point
    intercept = first_non_na_var - slope * first_non_na_series_position
    
    # Update the intercept of the model
    model.intercept_ = intercept
    
    # Predict and fill missing values in data
    missing_data_indices = data[data[var].isna() & (data['series_position'] < first_non_na_series_position)].index
    for index in missing_data_indices:
        series_position = data.loc[index, 'series_position']
        predicted_var = model.predict([[series_position]])
        data.loc[index, var] = predicted_var

    return data

# Define function to linearly interpolate data forcing intercept through last available data
def linearly_interpolate_after_end(data, var):
    
    # Identify maxmimum series position for which population data is available
    last_non_na_index = data[var].last_valid_index()
    last_non_na_series_position = data.loc[last_non_na_index, 'series_position']
    last_non_na_var = data.loc[last_non_na_index, var]
    
    # Drop rows with missing population values
    model_data = data.dropna(subset=[var])

    # Select last ten data points
    model_data = model_data[-10:]
    
    # Extract features and target
    X = model_data['series_position'].values.reshape(-1, 1)
    y = model_data[var].values.reshape(-1, 1)
    
    # Fit linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Calculate the slope of the regression line
    slope = model.coef_[0][0]
    
    # Calculate the intercept to force the regression line through the desired point
    intercept = last_non_na_var - slope * last_non_na_series_position
    
    # Update the intercept of the model
    model.intercept_ = intercept
    
    # Predict and fill missing values in data
    missing_data_indices = data[data[var].isna() & (data['series_position'] > last_non_na_series_position)].index
    for index in missing_data_indices:
        series_position = data.loc[index, 'series_position']
        predicted_var = model.predict([[series_position]])
        data.loc[index, var] = predicted_var

    return data

# Define function to linearly interpolate data forcing intercept between available data
def linearly_interpolate_between(data, var):
    
    # Ensure the data is sorted by the series position
    data = data.sort_values(by='series_position')
    
    # Linearly interpolate missing values for the specified column
    data[var] = data[var].interpolate(method='linear', limit_direction='both')
    
    return data

# Define function to impute by group averages and flag imputations
def impute_missing_values_with_flags(df, cols):
    # Group the DataFrame by 'cma_id' and 'month'
    grouped = df.groupby(['cma_id', 'month'])
    
    # Iterate over each column that needs imputation
    for col in cols:
        # Create a flag column to indicate where imputations happen
        flag_col = col + '_impute_flag'
        
        # Create a boolean series indicating where imputations are needed
        df[flag_col] = df[col].isna().astype(int)
        
        # Fill NA values with the mean of the group
        df[col] = grouped[col].transform(lambda x: x.fillna(x.mean()))
    
    return df

# Extract the last four digits of the CMA ID
price_index['cma_id_last4'] = price_index['cma_id'].str[-4:]
population['cma_id_last4'] = population['cma_id'].str[-4:]
construction_investment['cma_id_last4'] = construction_investment['cma_id'].str[-4:]

# Merge data
df = pd.merge(price_index, population[['year', 'month', 'population', 'cma_id_last4']], how='left', on=['cma_id_last4', 'year', 'month'])
df = pd.merge(df, construction_investment.drop(columns=['cma_id']), how='left', on=['cma_id_last4', 'year', 'month'])

# Drop the temporary columns used for merging
df = df.drop(columns=['cma_id_last4'])

# Apply the linear interpolation function to linearly trending data
df = df.groupby('cma').apply(lambda group: linearly_interpolate_before_start(group, 'population')).reset_index(drop=True)
df = df.groupby('cma').apply(lambda group: linearly_interpolate_after_end(group, 'population')).reset_index(drop=True)
df = df.groupby('cma').apply(lambda group: linearly_interpolate_between(group, 'population')).reset_index(drop=True)

# Apply the group mean imputation function to non-trending data
cols = ['new_construction_apartment', 'new_construction_row', 'new_construction_single',
        'types_of_work_apartment', 'types_of_work_row', 'types_of_work_single', 'types_of_work_total_residential']
df = impute_missing_values_with_flags(df, cols)
```

### Build Model

```python
# Normalize the data
for col in df.iloc[:,7:].columns:
    scaler = MinMaxScaler()
    df[col] = scaler.fit_transform(df[[col]])

# Define sequence length
seq_length = 24

def create_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data.iloc[i:i+seq_length]
        target = data.iloc[i+seq_length]['price_index']
        sequences.append(seq[['population', 'price_index']].values)
        targets.append(target)
    return np.array(sequences), np.array(targets)

all_sequences = []
all_targets = []
all_test_sequences = []
all_test_targets = []

# Prepare sequences for each CMA
for cma in df['cma'].unique():
    cma_data = df[df['cma'] == cma].sort_values(by='series_position')
    
    # Split into train and test
    train_data = cma_data.iloc[:-24]
    test_data = cma_data.iloc[-24-seq_length:]
    
    train_sequences, train_targets = create_sequences(train_data, seq_length)
    test_sequences, test_targets = create_sequences(test_data, seq_length)
    
    all_sequences.append(train_sequences)
    all_targets.append(train_targets)
    all_test_sequences.append(test_sequences)
    all_test_targets.append(test_targets)

# Combine sequences from all CMAs
X_train = np.concatenate(all_sequences, axis=0)
y_train = np.concatenate(all_targets, axis=0)
X_test = np.concatenate(all_test_sequences, axis=0)
y_test = np.concatenate(all_test_targets, axis=0)

# Positional encoding layer
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        self.d_model = d_model

        pe = np.zeros((max_len, d_model))
        position = np.arange(0, max_len)[:, np.newaxis]
        div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
        pe[:, 0::2] = np.sin(position * div_term)
        pe[:, 1::2] = np.cos(position * div_term)
        pe = pe[np.newaxis, ...]

        self.positional_encoding = tf.constant(pe, dtype=tf.float32)

    def call(self, inputs):
        seq_len = tf.shape(inputs)[1]
        return inputs + self.positional_encoding[:, :seq_len, :]

# Transformer model
def create_transformer_model(seq_length, num_features, d_model=128, num_heads=8, dff=512, num_layers=4):
    inputs = Input(shape=(seq_length, num_features))

    # Project inputs to d_model dimensions
    x = Dense(d_model)(inputs)

    # Add positional encoding
    x = PositionalEncoding(d_model)(x)

    # Transformer Encoder
    for _ in range(num_layers):
        x_prev = x
        x = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(x, x)
        x = Dropout(0.1)(x)
        x = LayerNormalization(epsilon=1e-6)(x + x_prev)
        x_prev = x
        x = Dense(dff, activation='relu')(x)
        x = Dense(d_model)(x)
        x = LayerNormalization(epsilon=1e-6)(x + x_prev)

    x = GlobalAveragePooling1D()(x)
    outputs = Dense(1)(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    return model

# Create the transformer model
model = create_transformer_model(seq_length, X_train.shape[2])

# Train the model
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Inverse transform predictions and actual values
y_test_inv = scaler_price.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler_price.inverse_transform(y_pred)

mse_inv = mean_squared_error(y_test_inv, y_pred_inv)
print(f'Mean Squared Error (Inverse Transformed): {mse_inv}')

# Plot the results for a sample of CMAs
sample_cmas = df['cma'].unique()[:3]  # Select a few CMAs to plot

for cma in sample_cmas:
    cma_data = df[df['cma'] == cma].sort_values(by='series_position')
    test_data = cma_data.iloc[-24-seq_length:]
    
    test_sequences, _ = create_sequences(test_data, seq_length)
    y_true = scaler_price.inverse_transform(test_data['price_index'].values[-24:].reshape(-1, 1))
    y_pred = model.predict(test_sequences)

    plt.figure(figsize=(10, 6))
    plt.plot(y_true, label='Actual Housing Prices')
    plt.plot(y_pred, label='Predicted Housing Prices')
    plt.title(f'Actual vs Predicted Housing Prices for CMA: {cma}')
    plt.xlabel('Time')
    plt.ylabel('Housing Prices')
    plt.legend()
    plt.show()
```
